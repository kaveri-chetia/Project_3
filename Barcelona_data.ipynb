```python
import pandas as pd
```

```python
listings_1 = pd.read_csv('Barcelona_listing_q1_2024.csv') #listings Q1 2024
listings_4 = pd.read_csv('Barcelona_listing_q2_2023.csv') #listings Q2 2023
listings_3 = pd.read_csv('Barcelona_listing_q3_2023.csv') #listings Q3 2023
listings_2 = pd.read_csv('Barcelona_listing_q4_2023.csv') #listings Q4 2023
```

```python
#Adding a new column to each dataframe to identify the quarter
listings_1['quarter'] = 'Q1_2024'
listings_2['quarter'] = 'Q4_2023'
listings_3['quarter'] = 'Q3_2023'
listings_4['quarter'] = 'Q2_2023'
```

```python
listings_df = pd.concat([listings_1, listings_2, listings_3,listings_4],ignore_index=True, axis=0)
listings_df
listings_df.shape
```
