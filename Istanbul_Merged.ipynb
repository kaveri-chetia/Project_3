```python
import pandas as pd

df = pd.read_csv('Istanbul_Q124_Listings.csv')
```

```python
ist_listings_1 = pd.read_csv('Istanbul_Q124_Listings.csv')
ist_listings_2 = pd.read_csv('Istanbul_Q423_Listings.csv')
ist_listings_3 = pd.read_csv('Istanbul_Q323_Listings.csv')
ist_listings_4 = pd.read_csv('Istanbul_Q223_Listings.csv')
```

```python
ist_listings_1.isnull().sum()
```

```python
#Adding a new column to each dataframe to identify the quarter
ist_listings_1['quarter'] = 'Q1_2024'
ist_listings_2['quarter'] = 'Q4_2023'
ist_listings_3['quarter'] = 'Q3_2023'
ist_listings_4['quarter'] = 'Q2_2023'
```

```python
ist_listings__merged = pd.concat([ist_listings_1, ist_listings_2, ist_listings_3, ist_listings_4],ignore_index=True, axis=0)
```

```python
ist_listings__merged
```

```python
ist_listings__merged.duplicated().sum() #0 duplicates
```

```python
columns_to_remove = ['last_review', 
                    'number_of_reviews_ltm', 
                    'license', 
                    'neighbourhood_group']

ist_listings__merged.drop(columns=columns_to_remove, inplace=True)
```

```python
ist_listings__merged.isnull().sum()
```

```python
# Calculate the mean of the 'price' column (ignoring NaN values)
mean_price = ist_listings__merged['price'].mean().round(2)
```

```python
# Fill NaN values in the 'price' column with the mean price
ist_listings__merged['price'] = ist_listings__merged['price'].fillna(mean_price).round(2)
```
